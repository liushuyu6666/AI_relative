```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython import display
import time
```

check if vector in the plane defined by the set of vectors
===================

```
# matrix S
S = np.array([[3, 0],
             [5, 2],
             [1, 2]])

# vector v
v1 = np.array([[-3], [1], [5]])
v2 = np.array([[1], [7], [3]])
```

cross product cp, which is a vector that is perpendicular to both a and b, and thus normal to the plane containing them
$$ A \times B = ||A|| \space ||B|| sin\theta n$$

```
cp = np.cross(S[:,0], S[:,1])
cp
```

define the plane

```
# draw plane corresponding to the column space
xx, yy = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
z1 = (-cp[0]*xx -cp[1]*yy)/cp[2]
z1
```

```
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot_surface(xx, yy, z1, alpha=0.2)

## plot the two vectors from matrix S
ax.plot([0, S[0, 0]], [0, S[1,0]], [0, S[2,0]], 'k')
ax.plot([0, S[0, 1]], [0, S[1,1]], [0, S[2,1]], 'k')

# and the vector v
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'r')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r')

ax.view_init(elev=1, azim=0)
plt.show()
```

change the equations system but have the same solution
================
2D

```
# ay = bx + c
eq1o = [1., 2, 1] # [a, b, c]
eq2o = [2., 1, 3]

for i in range(10):

    # clear plot
    # plt.cla()

    # randomly update equations
    eq1 = np.add(eq2o, np.random.randn(1)*eq1o)
    eq2 = np.add(eq1o, np.random.randn(1)*eq2o)

    # plot new lines
    y = ([eq1[1]*-3, eq1[1]*3] + eq1[2]) / eq1[0]
    plt.plot([-3, 3],y)

    y = ([eq2[1]*-3, eq2[1]*3] + eq2[2]) / eq2[0]
    plt.plot([-3,3],y)
    plt.axis([-3,3,-3,6])

    # pause to allow inspection
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(.1)
```

3D

```
# az = bx + cy + d
eq1o = [1, 2, 3, -1]
eq2o = [2, 1, 3, 3]

# set up for 3d plot
fig = plt.figure()
ax = fig.gca(projection='3d')

for i in range(0, 20):
    # plt.cla()

    # randomly update equations
    eq1 = np.add(eq2o, np.random.randn(1)*eq1o)
    eq2 = np.add(eq1o, np.random.randn(1)*eq2o)

    # plot new lines (solutions)
    y = ([eq1[1]*-3, eq1[1]*3] + eq1[3]) / eq1[0]
    z = ([eq1[2]*-3, eq1[2]*3] + eq1[3]) / eq1[0]
    ax.plot([-3, 3], y, z)

    # plot new lines (solutions)
    y = ([eq2[1]*-3, eq2[1]*3] + eq2[3]) / eq2[0]
    z = ([eq2[2]*-3, eq2[2]*3] + eq2[3]) / eq2[0]
    ax.plot([-3, 3], y, z)

    # axis limits
    ax.set_xlim3d(-3,6)
    ax.set_ylim3d(-3,6)
    ax.set_zlim3d(-1,10)

    # pause to allow inspection
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(.1)
```

all lines lay on the same plane  
This is not the same as Reduced row echelon form, which change the plane  
So we we can see from 2-D and 3-D figure is that this kind of transformation will not chagne the solution of this matrix.  
$$R^{new}_1 = a_1R^{old}_1 + b_1R^{old}_2 + c_1R^{old}_3 + ...$$
$$R^{new}_2 = a_2R^{old}_1 + b_2R^{old}_2 + c_2R^{old}_3 + ...$$
$$R^{new}_3 = a_3R^{old}_1 + b_3R^{old}_2 + c_3R^{old}_3 + ...$$

Reduced row echelon form
===============
Just help you to find the solution, but new vectors will in the different plane.

```
from sympy import Matrix

A = np.array([[3, 0],
             [5, 2],
             [1, 2]])
rrefA = np.array(Matrix(A).rref()[0])

rrefA = np.array([[1, 0],
             [0, 1],
             [0, 0]])
```

now, let's plot, we need to get cross vector

```
cp1 = np.cross(A[:,0], A[:,1])
cp2 = np.cross(rrefA[:,0], rrefA[:,1])
```

then define the plane

```
xx1, yy1 = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 11))
z1 = (-cp1[0]*xx1 -cp1[1]*yy1)/cp1[2]
xx2, yy2 = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 11))
z2 = (-cp2[0]*xx2 -cp2[1]*yy2)/cp2[2]
```

```
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(xx1, yy1, z1, alpha=0.2, color='red')

fig = plt.figure()
ax2 = fig.gca(projection='3d')
ax2.plot_surface(xx2, yy2, z2, alpha=0.2, color='blue')

## plot the two vectors from matrix A
ax.plot([0, A[0, 0]], [0, A[1,0]], [0, A[2,0]], 'k')
ax.plot([0, A[0, 1]], [0, A[1,1]], [0, A[2,1]], 'k')

# and the vector from rrefA
ax2.plot([0, rrefA[0, 0]], [0, rrefA[1,0]], [0, rrefA[2,0]], 'r')
ax2.plot([0, rrefA[0, 1]], [0, rrefA[1,1]], [0, rrefA[2,1]], 'r')

ax.view_init(elev=0, azim=4)
ax2.view_init(elev=50, azim=4)
plt.show()
```

you can see, two vectors systems is not in the same plane, but the intersection is the same.

any square full rank matrix has the identity matrix as its reduced row echelon form
