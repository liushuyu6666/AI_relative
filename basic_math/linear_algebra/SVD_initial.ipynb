# SVD

```
U, S, V = np.linalg.svd(pic) # convert pic to np.array first
plt.plot(S, 's-')
plt.xlim([0, 50]) # x limitation
plt.xlabel('Component number')
plt.ylabel('Singular value (\sigma)')
plt.show()
```

Reconstruct the image by using the several most important components


```
bDim = 0
eDim = 10
comps = np.arange(bDim, eDim)
comps
```

```
reconPic = U[:, comps]@np.diag(S[comps])@V[comps,:]
reconPic
```

```
plt.figure(figsize=(5, 10))
plt.subplot(1, 2, 1)
plt.imshow(pic, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(reconPic, cmap='gray')
plt.title('Reconstruct')
plt.axis('off')

plt.show()
```
